```python
import pandas as pd
```

```python
# input csv name
csv = 'interpolated_meter_data.csv'
```

```python
df = pd.read_csv(csv, encoding='utf-8')
```

```python
# make dataframe contain only interpolated data
df = df[df['is_exact'] == True]
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
df.to_csv('meter_data_only_interpolated.csv', index=False)
```

```python
print(df)
```

```python
meter_counts = df.groupby('meter_name').size()

print(f'Records per meter: {meter_counts}')
```

```python
meter_df = df.groupby('meter_name').size().reset_index(name='counts')
#meter_df.to_csv('meter_counts
```

```python
ls = []

for meter in df['meter_name'].unique():
    meter_data = df[df['meter_name'] == meter]

    start = meter_data['datetime'].iloc[0]
    end = meter_data['datetime'].iloc[-1]
    actual_count = len(meter_data)

    time_difference = (end - start).total_seconds() / 60
    expected_count = int(time_difference / 15) + 1

    ls.append({
        'meter_name': meter,
        'gaps_exist': actual_count != expected_count,
        'missing_num_rows': expected_count - actual_count,
        'expected': expected_count,
        'actual': actual_count,
        'date_range': f'{start} to {end}',
    })

count_df = pd.DataFrame(ls)
        
```

```python
total_meters = df['meter_name'].nunique()
missing_count = (count_df['gaps_exist'] == True).size
print(f"Total meters checked: {total_meters}")
print(f"Total meters missing timestamps: {missing_count}")
```

```python
print(count_df)
```

```python
count_df.to_csv('meter_counts.csv', index=False)
```
