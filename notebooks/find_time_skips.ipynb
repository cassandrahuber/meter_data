```python
import pandas as pd
```

```python
# input csv name
csv = 'interpolated_meter_data.csv'
```

```python
df = pd.read_csv(csv, encoding='utf-8')
```

```python
df = df[df['is_exact'] == True]
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
print(df)
```

```python
meter_counts = df.groupby('meter_name').size()

print(f'Records per meter: {meter_counts}')
```

```python
missing_count = 0
# checking time gaps
for meter in df['meter_name'].unique():
    meter_data = df[df['meter_name'] == meter]
    
    start = meter_data['datetime'].min()
    end = meter_data['datetime'].max()
    actual_count = len(meter_data)

    time_difference = (end - start).total_seconds() / 60
    expected_count = int(time_difference / 15) + 1

    if actual_count != expected_count:
        print(f'\nMeter: {meter}')
        print(f'   Missing = {expected_count - actual_count}')
        print(f'   Expected: {expected_count}, Actual: {actual_count}')
        print(f'   Date range: {start} to {end}')
        missing_count += 1

        # find the missing expected time intervals
```

```python
print(f"Total meters checked: {df['meter_name'].nunique()}")
print(f"Total meters missing timestamps {missing_count}")
```
