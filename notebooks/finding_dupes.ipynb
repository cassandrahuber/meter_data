```python
import os
import pandas as pd
```

```python
base_path = '/Volumes/Untitled/MeterDataTest'

if not os.path.exists(base_path):
    print("Error path does not exist")
    exit(1)
```

```python
df_list = []
```

```python
for sub_folder in os.listdir(base_path):
    sub_folder.replace(' ', '')
    folder_path = os.path.join(base_path, sub_folder)
    
    meter_name = sub_folder.lower().replace(' ', '_')

    csv_paths = [os.path.join(folder_path, f)
                 for f in os.listdir(folder_path)
                 if f.endswith('.csv')
                 and not f.startswith('._')
                 and not f.startswith('.')
                ]
    if meter_name == 'admin_serv_1_mtr':
        for csv in csv_paths:
            df = pd.read_csv(csv, encoding='utf-8')
            df.columns = df.columns.str.strip().str.replace(' ', '_')
            df_list.append(df)
            
            
```

```python
ad_df = pd.concat(df_list, ignore_index=True)
```

```python
dupes = ad_df[ad_df.duplicated(keep=False)]
```
